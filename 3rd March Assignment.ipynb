### 1.

**Any five plots that we are using in Seaborn libraries are scatterplot, catplot, pairplot, lineplot, displot.**

```python
import seaborn as sns
```

```python
a = sns.load_dataset('tips')
```

```python
a
```

```python
sns.scatterplot(x = 'total_bill', y = 'tip', data = a)
```

```python
sns.lineplot(x = 'total_bill', y = 'tip', data = a)
```

```python
sns.catplot(x = 'smoker', y = 'tip', data=a)
```

```python
sns.pairplot(a)
```

```python
sns.displot(a['smoker'])
```

### 2.

```python
b = sns.load_dataset("fmri")
```

```python
sns.lineplot(x ="timepoint", y = "signal", hue="event", style="region", data=b)
```

### 3.

```python
c = sns.load_dataset("titanic")
```

```python
sns.boxplot(x ='pclass', y = 'age', data=c)
```
